# **SQL Prompts and Examples:**

Prompt: Create the books and members tables to track inventory and library membership, then insert sample rows.

```python
import sqlite3
import pandas as pd

# Connect to an in-memory SQLite database
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

# Create a sample table: Products
cursor.execute(
    """
CREATE TABLE products (
    id INTEGER PRIMARY KEY,
    product_name TEXT NOT NULL,
    category TEXT NOT NULL,
    price REAL NOT NULL,
    stock_quantity INTEGER NOT NULL,
    launch_date DATE
);
"""
)

# Insert sample data
products_data = [
    (1, "Wireless Mouse", "Electronics", 25.50, 120, "2021-03-15"),
    (2, "Running Shoes", "Apparel", 85.00, 45, "2020-08-22"),
    (3, "Mechanical Keyboard", "Electronics", 75.00, 60, "2022-01-10"),
    (4, "Coffee Maker", "Home Appliances", 45.99, 30, "2019-11-05"),
    (5, "Bluetooth Speaker", "Electronics", 55.00, 85, "2021-07-19"),
    (6, "Leather Wallet", "Accessories", 35.00, 150, "2018-05-12"),
    (7, "Air Fryer", "Home Appliances", 120.00, 20, "2022-04-01"),
    (8, "Backpack", "Accessories", 40.00, 95, "2020-10-14"),
    (9, "Smart Watch", "Electronics", 150.00, 40, "2022-06-25"),
    (10, "Winter Jacket", "Apparel", 110.00, 25, "2021-11-30"),
]

cursor.executemany(
    "INSERT INTO products VALUES (?, ?, ?, ?, ?, ?)", products_data
)
conn.commit()
print("Sample 'products' table created and populated successfully.")
```

# **Prompt 1:**
 Select all products
Retrieve all records and columns stored within the books catalog table.

```python
sql_query = "SELECT * FROM products;"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

# **Prompt 2:**
 Select products from a specific category

```python
sql_query = "SELECT product_name, price, stock_quantity FROM products WHERE category = 'Electronics';"

df = pd.read_sql_query(sql_query, conn)
display(df)
```


# **Prompt 3:**
 Order products by price

```python
sql_query = "SELECT * FROM products ORDER BY price DESC;"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

# **Prompt 4:**
Count products per category

```python
sql_query = "SELECT category, COUNT(id) AS num_products FROM products GROUP BY category;"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

# **Prompt 5:**
Calculate average price per category

```python
sql_query = "SELECT category, AVG(price) AS average_price FROM products GROUP BY category;"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

# **Prompt 6:**
 Find products with price above a certain amount
Python

```python
sql_query = "SELECT product_name, category, price FROM products WHERE price > 50 ORDER BY price DESC;"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

# **Prompt 7:**
Find products launched after a specific date

```python
sql_query = "SELECT product_name, category, launch_date FROM products WHERE launch_date > '2021-01-01' ORDER BY launch_date ASC;"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

# **Prompt 8:**
 Find the highest price in each category

```python
sql_query = "SELECT category, MAX(price) AS max_price FROM products GROUP BY category;"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

# **Prompt 9:**
Products whose names start with a specific letter
Python

```python
sql_query = "SELECT product_name, category FROM products WHERE product_name LIKE 'W%';"

df = pd.read_sql_query(sql_query, conn)
display(df)
```

# **Prompt 10:**
Products with price within a range

```python
sql_query = "SELECT product_name, category, price FROM products WHERE price BETWEEN 30 AND 80 ORDER BY price;"

df = pd.read_sql_query(sql_query, conn)
display(df)
conn.close()
```
