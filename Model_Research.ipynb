# **Supervised Classification Options**
**1. Logistic Regression**
How it works: Evaluates basic input parameters to execute straightforward binary classifications (Normal vs. Attack).

Pros & Cons: Extremely fast to train and mathematically simple, but fails to map complex, multi-layered exploit patterns.

Best Used For: Filtering basic spam footprints or classifying simple, predictable security flags.

**2. Decision Trees**
How it works: Segments traffic features through an easily traceable, hierarchical rule structure.

Pros & Cons: Offers excellent structural clarity and requires minimal data cleaning, but overfits easily on noisy datasets.

Best Used For: Map-driven incident classification and basic rule sorting workflows.

**3. Random Forest**
How it works: Bundles vast arrays of independent decision trees to boost prediction reliability.

Pros & Cons: Delivers strong classification accuracy and handles high feature counts well, but demands significant memory and slows down during heavy real-time tracking.

Best Used For: Signature-heavy network intrusion detection systems.

**4. XGBoost (Extreme Gradient Boosting)**
How it works: Builds sequential, error-correcting decision trees optimized for maximum prediction speed and accuracy.

Pros & Cons: Exceptional accuracy on structured tabular logs and robust handling of imbalanced datasets, though it requires extensive hyperparameter tuning.

Best Used For: Categorizing known attack variants and pinpointing signature risks.

# **Unsupervised Outlier Options**
**5. Isolation Forest**

How it works: Isolates anomalies directly by randomly partitioning data points; outliers end up with drastically shorter paths in the trees.

Pros & Cons: Exceptional at spotting zero-day exploits without requiring any pre-labeled data, but cannot classify the exact flavor of the attack on its own.

Best Used For: Insider threat profiling and catching abnormal network traffic deviations.
