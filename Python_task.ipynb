# Prompt 1:

Goal: Simulate loading a dataset of security logs that would be used for threat detection. For demonstration purposes, we'll create a synthetic dataset.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

CORE PYTHON: STUDENT REGISTRY & GRADE LOOKUP

# **Prompt-2**
Tuple for fixed student roster: (Student_ID, Student_Name)

```python
student_roster = [("S01", "Alice"), ("S02", "Bob"), ("S03", "Charlie"), ("S04", "Diana")]
```

# **Prompt 3**
Dictionary for exam score lookups: {Student_ID: Test_Score}

```python
math_scores = {"S01": 85, "S02": 58, "S03": 92, "S04": 74}
def get_grade_status(score):
    """Checks if a student is passing or needs extra help."""
    if score >= 60:
        return "Passing"
    else:
        return "Needs Help"
```

# **Prompt 4-**
NUMPY: WEEKLY ATTENDANCE TRACKER

1D Array tracking days attended this month by the 4 students (Max 20 days)

```python
attendance_days = np.array([19, 15, 20, 18])
print("=== CLASS ATTENDANCE METRICS ===")
print(f"Total classroom desks filled this month: {np.sum(attendance_days)} days")
print(f"Average days attended per student: {np.mean(attendance_days):.1f} days")
print(f"Perfect attendance record: {np.max(attendance_days)} days\n")
```

# **Prompt 5-**
PANDAS: REPORT CARD GENERATOR

```python
classroom_data = {
    "Student_ID": [student[0] for student in student_roster],
    "Student_Name": [student[1] for student in student_roster],
    "Attendance_Days": attendance_days.tolist(),
}

df_classroom = pd.DataFrame(classroom_data)
```

# **Prompt 6-**
Map our core math scores and statuses into the dataframe

```python

df_classroom["Math_Score"] = df_classroom["Student_ID"].map(math_scores)
df_classroom["Status"] = df_classroom["Math_Score"].apply(get_grade_status)

print("=== CLASSROOM PERFORMANCE REPORT ===")
print(df_classroom.to_string(index=False))
```

# **Prompt 7-**
Filter out students who need extra help (Score < 60)

```python

students_struggling = df_classroom[df_classroom["Math_Score"] < 60]
print("\n=== STUDENTS REQUIRING TUTORING ===")
print(students_struggling[["Student_Name", "Math_Score", "Status"]].to_string(index=False))
```

# **Prompt 8-**
VISUALIZATION: GRADE VISUALIZER

```python
plt.figure(figsize=(7, 4))
sns.barplot(
    data=df_classroom,
    x="Student_Name",
    y="Math_Score",
    hue="Student_Name",
    palette="muted",
    legend=False,
)
plt.axhline(60, color="red", linestyle="--", label="Passing Mark (60)")
plt.title("Math Exam Scores by Student")
plt.xlabel("Student Name")
plt.ylabel("Exam Score")
plt.ylim(0, 100)
plt.legend()
plt.tight_layout()
plt.show()
```

# **prompt 9-**
plot the status of distribution


```python
def plot_status_distribution(dataframe):
    """Generates a pie chart displaying passing vs. struggling class breakdowns."""
    status_counts = dataframe["Status"].value_counts()
    plt.figure(figsize=(5, 5))
    plt.pie(
        status_counts,
        labels=status_counts.index,
        autopct="%1.1f%%",
        startangle=140,
        colors=["#66b3ff", "#ff9999"],
    )
    plt.title("Classroom Academic Status Breakup")
    plt.tight_layout()
    plt.show()
plot_status_distribution(df_classroom)
```

# **Prompt 10-**
plotting attendance vs scores

```python
def plot_attendance_vs_scores(dataframe):
    """Generates a scatter plot analyzing the correlation between tracking metrics."""
    plt.figure(figsize=(6, 4))
    sns.scatterplot(
        data=dataframe,
        x="Attendance_Days",
        y="Math_Score",
        hue="Status",
        s=200,
        palette={"Passing": "green", "Needs Help": "red"},
    )
    plt.title("Impact of Attendance on Final Exam Scores")
    plt.xlabel("Days Attended (Out of 20)")
    plt.ylabel("Math Exam Score")
    plt.xlim(10, 22)
    plt.ylim(40, 100)
    plt.grid(True, linestyle=":", alpha=0.6)
    plt.tight_layout()
    plt.show()

plot_attendance_vs_scores(df_classroom)
```
